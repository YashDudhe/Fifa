```python
def year(x):
    l=[]
    if isinstance(x,str):
        l=x.split(", ")
        x=l[-1]
    return(x)

df['Joined']=df['Joined'].replace(np.nan,0)
df['Joined'] =df['Joined'].apply(year).astype('str')
```

```python
import datetime
df['Contract Valid Until']=pd.to_datetime(df['Contract Valid Until'])
df['year'] = pd.DatetimeIndex(df['Contract Valid Until']).year.astype(str)
df['Contract Valid Until'].fillna('Not available', inplace = True) 
```

```python
def clean_height(x):
    l=[]
    s=1
    if isinstance(x,str):
        l=x.split("'")
        i=int(l[0])
        j=int(l[1])
        s=((i*12)+j)/12
    return(s)


df['Height'] = df['Height'].apply(clean_height)
df['Height'].fillna((df['Height'].mean()), inplace = True)
```

```python
def clean_weight(x):
    
    if isinstance(x,str):
        return(x.replace('lbs', ''))
    return(x)

df['Weight'] = df['Weight'].apply(clean_weight).astype('float')
df['Weight'].fillna((df['Weight'].mean()), inplace = True) 
```

```python
def skillConverter(val):
    if type(val) == str:
        s1 = val[0:2]
        s2 = val[-1]
        val = int(s1) + int(s2)
        return val
    else:
        return val

skill_columns = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

for col in skill_columns:
    df[col] = df[col].apply(skillConverter)
    df[col].fillna(0.0, inplace = True)
```

```python
def conv(num):
    if num[-1]=='M':
        return float(num[1:-1])*1000000
        
    elif num[-1]=='K':
        return float(num[1:-1])*1000
    else:
        return float(num[1:])


fifa['Value'] = fifa['Value'].apply(lambda x: conv(x))
fifa['Wage'] = fifa['Wage'].apply(lambda x: conv(x))
```

## COMPARING TWO PLAYERS 

##                                            MESSI VS RONALDO

```python
skills=[
    'Crossing', 'Finishing', 'HeadingAccuracy',
       'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed',
       'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping',
       'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions',
       'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking',
       'StandingTackle', 'SlidingTackle'
]
```

```python
messi=fifa.loc[fifa['Name']=='L. Messi']
messi= pd.DataFrame(messi,columns=skills)

ronaldo=fifa.loc[fifa['Name']=='Cristiano Ronaldo']
ronaldo= pd.DataFrame(ronaldo,columns=skills)
```

```python
messi
```

```python
ronaldo
```

```python
plt.figure(figsize=(40,10))
sns.pointplot(data=messi,color="red")
sns.pointplot(data=ronaldo,color="black")
plt.xlabel("SKILLS",fontsize = 30)
plt.ylabel("VALUE",fontsize = 30)
plt.title("MESSI VS RONALDO", fontsize = 50)

plt.grid()
```
